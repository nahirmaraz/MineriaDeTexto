```
from nltk.tokenize import word_tokenize

token = word_tokenize(text)
token
```

```
import string
new_words = []
for word in token:
    for letter in word:
        if letter in string.punctuation:
            word = word.replace(letter,"")
    new_words.append(word)

print(new_words)
```

```
new_list = []
for vacio in new_words:
  if vacio != "":
    new_list.append(vacio)
print(new_list)
```

```
minusculas=[]
for i in new_list:
    i = i.lower()
    minusculas.append(i)
print(minusculas)
```

```
from nltk import word_tokenize
from nltk.corpus import stopwords

nltk.download("stopwords")

a = set(stopwords.words('spanish'))
#print(a) #Muestra las palabras que son vacias para el ingles

stopwords = [x for x in minusculas if x not in a]
print(stopwords)
```
